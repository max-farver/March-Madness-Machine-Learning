```python
ncaaGames = pd.read_csv("..\Data\DataFiles\\NCAATourneyCompactResults.csv")
```

```python
ncaaGames = ncaaGames.drop(columns=['DayNum', 'WScore', 'LScore', 'WLoc', 'NumOT'], axis=1)
ncaaGames.head(10)
```

```python
def switch_vals(x):
    rand_val = random.randint(0,2)
    if rand_val == 0:
        return [x[0], x[2], x[1], x[2]]
    return [x[0], x[1], x[2], x[2]]
```

```python
ncaaGames['winner'] = 0
for i in range(len(ncaaGames)):
    ncaaGames.iloc[i] = switch_vals(ncaaGames.iloc[i])
ncaaGames.head(10)
```
