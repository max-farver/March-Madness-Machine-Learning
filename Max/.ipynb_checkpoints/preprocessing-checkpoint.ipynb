```python
import pandas as pd
import numpy as np
from  sklearn.preprocessing import LabelEncoder, StandardScaler


from sklearn.model_selection import  cross_val_score, cross_val_predict, train_test_split
from sklearn.metrics import r2_score, accuracy_score

from sklearn.externals import joblib

pd.options.display.max_columns = None
```

```python
data = pd.read_hdf("training_data.h5", key="df")
seeds = pd.read_csv("../Data/DataFiles/NCAATourneySeeds.csv")

seeds['Seed'] = seeds.apply(lambda x: x['Seed'].replace('W', ""), axis=1)
seeds['Seed'] = seeds.apply(lambda x: x['Seed'].replace('X', ""), axis=1)
seeds['Seed'] = seeds.apply(lambda x: x['Seed'].replace('Y', ""), axis=1)
seeds['Seed'] = seeds.apply(lambda x: x['Seed'].replace('Z', ""), axis=1)
seeds['Seed'] = seeds.apply(lambda x: x['Seed'].replace('a', ""), axis=1)
seeds['Seed'] = seeds.apply(lambda x: x['Seed'].replace('b', ""), axis=1)

data = pd.merge(data, seeds, left_on=['Season_x', 'TeamID_x'], right_on=['Season', 'TeamID'])
```

```python
data = pd.merge(data, seeds, left_on=['Season_y', 'TeamID_y'], right_on=['Season', 'TeamID'])
```

```python
data = data.drop(columns=['Season_x', 'Season_y', 'Season_TeamID_x', 'Season_TeamID_y'], axis=1)
data = data.loc[:,~data.columns.duplicated()]
```

```python
# encode categorical vals - all are nominal

# team_enc = LabelEncoder()
# team_enc.fit(data['winner'])
# data['winner'] = team_enc.transform(data['winner'])
Y = data['winner']
```

```python
# data['TeamID_x'] = team_enc.transform(data['TeamID_x'])
# data['TeamID_y'] = team_enc.transform(data['TeamID_y'])
```

```python
conf_enc = LabelEncoder()
conf_enc.fit(data['ConfAbbrev_x'])
data['ConfAbbrev_x'] = conf_enc.transform(data['ConfAbbrev_x'])
data['ConfAbbrev_y'] = conf_enc.transform(data['ConfAbbrev_y'])

joblib.dump(conf_enc, 'conf_enc_file.joblib')

seed_enc = LabelEncoder()
seed_enc.fit(data['Seed_x'])
joblib.dump(seed_enc, 'seed_enc_file.joblib')

data['Seed_x'] = seed_enc.transform(data['Seed_x'])
data['Seed_y'] = seed_enc.transform(data['Seed_y'])
X_cat = data[['ConfAbbrev_x','ConfAbbrev_y','TeamID_x','TeamID_y','Seed_x', 'Seed_y']]
data.head()
```

```python
# scaling quantitative vals
cols = data.columns
cols_quan = cols.drop(['ConfAbbrev_x','ConfAbbrev_y','winner','TeamID_x','TeamID_y', 'Seed_x', 'Seed_y'])
cols_quan
```

```python
scl = StandardScaler()
X_quan = data[cols_quan]
```

```python
scl.fit(X_quan)
joblib.dump(scl, 'scl_file.joblib')
```

```python
X_quan = scl.transform(X_quan)
```

```python
cleaned = pd.DataFrame(Y)
```

```python
cleaned = pd.concat([cleaned, pd.DataFrame(X_cat)], axis=1)
cleaned = pd.concat([cleaned, pd.DataFrame(X_quan)], axis=1)
```

```python
cleaned.head()
```

```python
cleaned['winner'] = cleaned.apply(lambda x: 0 if x['winner'] == x['TeamID_x'] else 1, axis=1)
```

```python
cleaned.head()
```

```python
cleaned.to_hdf("processed.h5", key="df", mode='w')
```

```python
Y = cleaned['winner']
X = cleaned.drop(columns=['winner', 'TeamID_x','TeamID_y'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42)
```
